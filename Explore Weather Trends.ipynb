```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('results.csv')
```

```python
df.head()
```

```python
# function for moving average
# for global_avg_temp
def moving_avg(mA_range, data_input):
    output = data_input.rolling(window = mA_range, on = "global_avg_temp").mean().dropna()
    return output
# Function Calling with the range of Moving Average
mA_value = 5
chart_moving_avg = moving_avg(mA_value, df)

# Drawing the graph: Global Temperature
plt.plot(chart_moving_avg ['year'], chart_moving_avg ['global_avg_temp'], label = 'Global')
plt.legend()
plt.xlabel ("Years")
plt.ylabel ("Temperature (°C)")
plt.title ("Global Average Temperature")
plt.show ()
```

```python
# for city_avg_temp
def moving_avg(mA_range, data_input):
    output = data_input.rolling(window = mA_range, on = "city_avg_temp").mean().dropna()
    return output
# Function Calling with the range of Moving Average
mA_value = 14
chart_moving_avg = moving_avg(mA_value, df)

# Drawing the graph: Global Temperature
plt.plot(chart_moving_avg ['year'], chart_moving_avg ['city_avg_temp'], label = 'City - Oakland')
plt.legend()
plt.xlabel ("Years")
plt.ylabel ("Temperature (°C)")
plt.title ("City Average Temperature")
plt.show ()
```

```python

plt.plot(chart_moving_avg ['year'], chart_moving_avg ['global_avg_temp'], label = 'Global',color ='r')
plt.plot(chart_moving_avg ['year'], chart_moving_avg ['city_avg_temp'], label = 'City - Oakland',color='y')
plt.legend(loc='bottomleft',bbox_to_anchor=(1.0, 1.0))
plt.xlabel ("Years")
plt.ylabel ("Temperature (°C)")
plt.title ("Average Temperature")
plt.show ()
```
